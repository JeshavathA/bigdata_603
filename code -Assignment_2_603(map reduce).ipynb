```python
from PyPDF2 import PdfReader
def ext(pdf_path, start_page, end_page):
    pdf_text = ""
    pdf_reader = PdfReader(pdf_path)
    for page_num in range(start_page - 1, end_page):
        page = pdf_reader.pages[page_num]
        pdf_text += page.extract_text()
    return pdf_text
pdf= 'Harry_Potter_(www.ztcprep.com).pdf'
start=4836
end=4853
e= ext(pdf,start,end)
with open('file1.txt','w',encoding='utf-8') as txt_file:
    txt_file.write(e)
print("Text extracted and saved to file1.txt")
```

```python
!pip install tabulate
import re
from collections import Counter
from tabulate import tabulate
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())
def mapper(text):
    w= tokenize(text)
    return Counter(w)
def reducer(counters):
    a= Counter()
    for i in counters:
        a.update(i)
    return a
with open('file1.txt','r',encoding='utf-8') as input_file:
    text=input_file.read()
c=len(text)//4
chunks=[]
for i in range(0,len(text),c):
    chunk= text[i:i+c]
    chunks.append(chunk)
mapped_results = []
for chunk in chunks:
    r= mapper(chunk)
    mapped_results.append(r)
ab=reducer(mapped_results)
data=[]
for i,j in ab.items():
    data.append((i,j))
print(tabulate(data, headers=["Word", "Count"], tablefmt="grid"))
```

```python
import re
from collections import Counter
from tabulate import tabulate
import spacy
def tokenize(text):
    return re.findall(r'"([^"]+)"|\S+',text.lower())
def is_space(word):
    return word.isspace()
def is_non_english(word):
    return any(ord(char) > 127 for char in word)
def mapper(text):
    w= tokenize(text)
    s_c= sum(1 for word in w if is_space(word))
    ne= [word for word in w if not is_space(word) and is_non_english(word)]
    nlp=spacy.load('en_core_web_sm')
    doc= nlp(text)
    entities= [ent.text for ent in doc.ents if ent.label_ in ['PERSON','GPE']]
    return Counter(ne + entities), s_c
def reducer(counters):
    nec= Counter()
    tsc=0
    for c,sc in counters:
        nec.update(c)
        tsc+=sc
    return nec,tsc
with open('file2.txt','r',encoding='utf-8') as input_file:
    t=input_file.read()
cs=len(t)//4
chs=[t[i:i+cs] for i in range(0, len(t), cs)]
mr=[mapper(ch) for ch in chs]
fnec,tsc=reducer(mr)
d=[(w,c) for w,c in fnec.items()]
print(tabulate(d, headers=["Names/Places","Count"], tablefmt="grid"))
```
